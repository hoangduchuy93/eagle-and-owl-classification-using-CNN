```python
#Part 3 - making new predictions
from tensorflow.keras.models import load_model 
classifier = load_model('model_CNN_predict_eagle_owl_20211106.h5')
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image 
```

```python
#đưa hình vào
test_image = image.load_img('image/single_prediction/bald_eagle.jpg',
                            target_size = (64, 64))
test_image = image.img_to_array(test_image)
print(test_image.shape)
test_image = test_image/255 #scale
test_image = np.expand_dims(test_image, axis = 0) #đưa ảnh vào
print(test_image.shape)

result = classifier.predict(test_image)
print(result)
if result[0][0] >= 0.5:
  prediction = 'owl'
else:
  prediction = 'eagle'

print(prediction)
```

```python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img=mpimg.imread('image/single_prediction/bald_eagle.jpg')
imgplot = plt.imshow(img)
plt.show()
```
